```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
```

```python
data = [['Milk', 'Onion', 'Nutneg', 'Kidney Beans','Eggs','Yogurt'],
        ['Dill','Onion', 'Nutneg', 'Kidney Beans','Eggs','Yogurt'],
        ['Milk', 'Apple', 'Kidney Beans', 'Eggs'],
        ['Milk','Unicorn', 'Corn','Kidney Beans', 'Yogurt'],
        ['Corn', 'Onion', 'Onion', 'Kidney Beans', 'Ice Cream', 'Eggs']]
data
```

```python
te = TransactionEncoder()
te_ary = te.fit(data).transform(data)
te
```

```python
df = pd.DataFrame(te_ary, columns = te.columns_)
df
```

```python
from mlxtend.frequent_patterns import apriori

frequent_patterns = apriori(df, min_support= 0.6, use_colnames= True)
frequent_patterns
```

# ASSociation Rule Mining

```python
from mlxtend.frequent_patterns import association_rules

res = association_rules(frequent_patterns, metric = "confidence", min_threshold = 0.7)
res
```

```python
res1 = res[['antecedents', 'consequents', 'support', 'confidence', 'lift']]
res1
```

```python
res2 = res1[res1['confidence'] >= 1]
res2
```
