```python
import pandas as pd 
import numpy as np
```

```python
data = pd.read_csv("linear-reggression.csv")
data
```

```python
duplicate = data.duplicated().sum()
duplicate
```

```python
import matplotlib.pyplot as plt
```

```python
data.describe()
```

```python
x = pd.DataFrame(data, columns = ["Happiness Score"])
y = pd.DataFrame(data, columns = ['Freedom'])

```

```python
plt.scatter(x,y, alpha = 0.3)
plt.xlabel("Happiness Score")
plt.ylabel("Freedom")

plt.xlim(0, 7.587000)
plt.ylim(0,0.669730)
```

```python
import sklearn
from sklearn.linear_model import LinearRegression
```

```python
linreg = LinearRegression()
linreg.fit(x,y)
```

```python
linreg.coef_
```

```python
linreg.intercept_
```

```python
plt.scatter(x,y, alpha = 0.3)
plt.xlabel("Happiness Score")
plt.ylabel("Freedom")
plt.plot(x, linreg.predict(x), color='yellow')

plt.xlim(0, 7.587000)
plt.ylim(0, 0.669730)
```
